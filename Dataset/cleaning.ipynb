```python
# For cleaing tool we will be using the tool called polars

import polars as pl 

datasets = pl.read_csv('datasests.csv')
```

```python
# To remov the null valus from the daatsets 

data = datasets.drop_nulls()
```

```python
# TO remeove the null valuses 

data = datasets.unique()
```

```python
# clean the text colums 

data = datasets.with_columns(
    pl.col('prompt_text').
    str.to_lowercase().
    str.replace_all(r'[^\w\s]', '').
    str.strip_chars('')
)
```

```python
# to remove the eompy words
datasets.filter(
    pl.col("prompt_text") != " "
)
```

```python
# sa etegh cleaing csv
data.write_csv('cleaned_datasets.csv')
```
